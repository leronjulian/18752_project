```python
from sklearn.decomposition import FastICA
import librosa as lb 
import matplotlib.pyplot as plt 
import librosa.display as libDisp
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
import pandas as pd
import numpy as np
```

```python
notes = []
for i in range(5, 7):
    for note in ['C', 'C#', 'D', 'D#', 'E', 'F', 'F#', 'G', 'G#', 'A', 'A#', 'B']:
        notes.append(f'{note}{i}')

mapping = {i: notes[i] for i in range(len(notes))}
```

```python
a = np.load('song1.npz', allow_pickle=True)

b = a['arr_0']
b = b.sum(axis=0)
```

```python
X = b['X'][:]
y = b['y'][:]

# Fast ICA
transformer = FastICA(n_components=2, random_state=0)
X_transformed = transformer.fit_transform(X)
```

```python
plt.scatter(X_transformed[:, 0], X_transformed[:, 1])
```

```python
def note_played_scatter(X, note_no, y):
    targets = y[:, note_no]
    colors = ['blue' if t == 1.0 else 'red' for t in targets ]
    plt.scatter(X[:, 0], X[:, 1], c=colors, marker='x', alpha=0.7)
    plt.title(f'note {mapping[note_no]} played (blue)')
    plt.show()
```

```python
for i in range(len(notes)):
    note_played_scatter(X_transformed, i, y)
```
